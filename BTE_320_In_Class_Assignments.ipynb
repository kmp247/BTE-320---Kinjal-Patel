Income Statement

```python
# Inputs: sales and cogs_pct and tax_rate (input = you inputting the number given after pressing play, float meaning a decimal output)
sales = input("Enter Sales ($): ")
sales = float(sales)

cogs_pct = input("Enter Cost of Goods sold (%): ")
cogs_pct = float(cogs_pct)

tax_rate = input("Enter Tax Rate (%): ")
tax_rate = float(tax_rate)

# Process: calculate actual cost of goods sold, gross profit, tax rate and netincome
cogs = sales * cogs_pct / 100
grossprofit = sales - cogs

taxrate = grossprofit * (tax_rate / 100)
netincome = grossprofit - taxrate

# output: display cogs and gross profit --> print can use several ways, using fstring
print(f"Cost of Goods Sold: ${cogs}")
print(f"Gross Profit: ${grossprofit}")
print(f"Tax Rate: ${taxrate}")
print(f"Net Income: ${netincome}")
```

```python
# Simple Numerical Calculator --> My attempt

# Inputs
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter the operator (+, -, *, /): ")

# Perform calculation
if operator == "+":
    result = num1 + num2
    print("Result:", result)

elif operator == "-":
    result = num1 - num2
    print("Result:", result)

elif operator == "*":
    result = num1 * num2
    print("Result:", result)

elif operator == "/":
    if num2 != 0:  # prevent division by zero
        result = num1 / num2
        print("Result:", result)
    else:
        print("Error: Division by zero is not allowed.")

else:
    print("Invalid operator. Please choose from +, -, *, /.")
```

Quality control manager (exceptions)

```python
def getValues(n: int) -> list:
    '''
    Prompts the user for inputs, adds them to a list
    Inputs:
    - Number of items in the list
    Output:
    - A list
    '''
    items = []
    for i in range(n):
        while True:
            try:
                item = float(input(f'Enter weight value #{i+1}: '))
                if item < 0:
                    raise ValueError("Weight cannot be negative.")
                items.append(item)
                break
            except ValueError as e:
                print(f"Invalid input: {e}. Please enter a positive number.")
    return items


def calcAverage(alist: list) -> float:
    '''
    Input:
    - a list of weights
    Output:
    - the average weight (excluding the first two values)
    '''
    if len(alist) <= 2:
        print("Not enough values to calculate average.")
        return float('nan')
    try:
        return sum(alist[2:]) / len(alist[2:])
    except Exception as e:
        print(f"Error calculating average: {e}")
        return float('nan')


def main():
    while True:
        try:
            n = int(input('How many weights? '))
            if n < 0:
                print("Number of weights cannot be negative.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter a whole number.")

    L = getValues(n)
    avg = calcAverage(L)
    print(f'Average (excluding first two): {avg:.2f} lbs.')


main()
```

Assignment: Loan & NPV mini-toolkit (exceptions)

```python
def roi(cost: float, revenue: float) -> float:
  assert cost > 0, "cost must be positive"
  assert isinstance(cost, (int, float)), "cost must be numerical"
  assert isinstance(revenue, (int, float)), "revenue must be numerical"

  return(revenue - cost) / cost

if __name__ == "_main_":
  c = float(input("Enter cost: "))
  r = float(input("Enter revenue: "))
  ret = roi(c, r)
  print(ret)
```
